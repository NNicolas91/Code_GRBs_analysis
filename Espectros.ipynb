```python
import matplotlib.pyplot as plt
import numpy as np

# Función para leer los datos
def leer_datos(archivo):
    datos = []
    with open(archivo, 'r') as file:
        grupo = []
        for linea in file:
            if 'NO NO NO NO NO' in linea:  # Si encontramos 'NO', añadimos el grupo actual a los datos y reiniciamos
                if grupo:
                    datos.append(grupo)
                    grupo = []
            else:
                valores = linea.strip().split()
                valores = [float(val) if val != 'NO' else np.nan for val in valores]
                grupo.append(valores)
        if grupo:  # Añadir el último grupo si hay datos
            datos.append(grupo)
    return datos

# Función para graficar los datos
def graficar_datos(datos):
    grupo_data = {}
    for i, grupo in enumerate(datos, start=1):
        grupo = np.array(grupo)
        x = grupo[:, 0]
        y = grupo[:, 2]
        errores_x = grupo[:, 1]
        errores_y = grupo[:, 3]
        modelo = grupo[:, 4]
        # Store the data for the current group in a dictionary
        grupo_data[i] = {
            'x': x,
            'y': y,
            'errores_x': errores_x,
            'errores_y': errores_y,
            'modelo': modelo
        }
    fig, axs = plt.subplots(2, figsize=(10, 6), constrained_layout=True, sharex=True)
    fig.suptitle('GRB070724A', fontsize=15)
    axs[0].errorbar(grupo_data[1]['x'], grupo_data[1]['y'], xerr=grupo_data[1]['errores_x'], yerr=grupo_data[1]['errores_y'], fmt='o',ms=1, color='black')
    axs[0].scatter(grupo_data[1]['x'], grupo_data[1]['y'], label='Modo WT', marker='*',color='black')
    axs[0].step(grupo_data[1]['x'], grupo_data[1]['modelo'], label='Model WT', linestyle = '--', color='black')
    axs[0].errorbar(grupo_data[2]['x'], grupo_data[2]['y'], xerr=grupo_data[2]['errores_x'], yerr=grupo_data[2]['errores_y'], fmt='o',ms=1, color='gray')
    axs[0].scatter(grupo_data[2]['x'], grupo_data[2]['y'], label='Modo PC', marker='o', s = 10, color='gray')
    axs[0].step(grupo_data[2]['x'], grupo_data[2]['modelo'], label='Model PC', color='gray',linestyle = '-.')
    axs[0].legend()
    axs[1].errorbar(grupo_data[3]['x'], grupo_data[3]['y'], xerr=grupo_data[3]['errores_x'], yerr=grupo_data[3]['errores_y'], fmt='o',ms=1, color='black')
    axs[1].axhline(0, color='gray', linestyle='-.')
    axs[1].errorbar(grupo_data[4]['x'], grupo_data[4]['y'], xerr=grupo_data[4]['errores_x'], yerr=grupo_data[4]['errores_y'], fmt='o',ms=1, color='gray')
    axs[0].set_yscale('log')
    axs[0].set_xscale('log')
    axs[1].set_xlabel('Energía $(keV)$', fontsize=16)
    axs[0].set_ylabel('Cuentas $s^{-1}$ $keV^{-1}$',  fontsize=16)
    axs[1].set_ylabel('Ratio',  fontsize=16)
    axs[0].grid(True)
    axs[1].grid(True)
    plt.savefig('GRB070724Axrtspectrapoisson.pdf')
    plt.show()

# Ruta al archivo de datos
archivo = '/content/pp070724A.txt'  # Cambiar al nombre de tu archivo

# Leer los datos
datos = leer_datos(archivo)

# Graficar los datos
graficar_datos(datos)
```
